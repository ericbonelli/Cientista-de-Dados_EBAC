#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo)
for ticker in ativos:
    yf_data[('amplitude', ticker)] = yf_data[('High', ticker)] - yf_data[('Low', ticker)]
yf_data
```

```python
#Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo., fundo branco
fig = graph_objects.Figure()
for ticker in ativos:
    fig.add_trace(graph_objects.Scatter(x=yf_data.index, y=yf_data[('amplitude', ticker)], name=ticker))
fig.show()
```

O Ibovespa apresenta a maior amplitude intraday
