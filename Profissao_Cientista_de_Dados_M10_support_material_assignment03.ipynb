1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
ax = sns.countplot(x='data_ref',  hue = 'sexo', data=data)

data['data_ref'] = pd.to_datetime(data['data_ref'])
tick_labs = data['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(data['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

**Não parece haver mudanças significativas na distribuição ao longo do tempo para variável Sexo.**

````python
ax = sns.countplot(x='data_ref',  hue = 'educacao', data=data)

data['data_ref'] = pd.to_datetime(data['data_ref'])
tick_labs = data['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(data['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

**As proporções das diferentes categorias de escolaridade parecem consistentes para variável Educação**

````python
ax = sns.countplot(x='data_ref',  hue = 'tipo_renda', data=data)

data['data_ref'] = pd.to_datetime(data['data_ref'])
tick_labs = data['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(data['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

**Não parece haver mudanças significativas na distribuição ao longo do tempo para variável Tipo de renda.**

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'sexo', data=data)

data['data_ref'] = pd.to_datetime(data['data_ref'])
tick_labs = data['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(data['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````

**As médias de renda para homens e mulheres parecem estáveis, sem alterações significativas**

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'educacao', data=data)

data['data_ref'] = pd.to_datetime(data['data_ref'])
tick_labs = data['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(data['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````

**As médias de renda por nível educacional são estáveis.**

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'tipo_renda', data=data)

data['data_ref'] = pd.to_datetime(data['data_ref'])
tick_labs = data['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(data['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=45)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

**As médias de renda por tipo de renda são consistentes.**

**Estabilidade:** As médias de renda para os níveis das variáveis qualitativas são majoritariamente estáveis no tempo.

**Aplicação em Modelagem:** Todas essas variáveis parecem adequadas para modelagem, pois têm padrões consistentes e podem agregar informação preditiva, especialmente se combinadas com outras variáveis
