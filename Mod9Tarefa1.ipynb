### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
data.groupby('munResNome')[['IDADEMAE', 'IDADEPAI']].mean()
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
aniversario = "09-16"
data['DTNASC'] = pd.to_datetime(data['DTNASC'])
filtered_data = data[data['DTNASC'].dt.strftime('%m-%d') == aniversario]
filtered_data.groupby(['ESCMAE', 'SEXO'])['PESO'].mean()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
# Qual o municipio que nasceu menos bebe em 2019
data.groupby('munResNome')['IDADEMAE'].count().sort_values(ascending=True)
```

```python
# qual a idade media, maxima, minima das maes nesse municipio?
filtered_data = data[data['munResNome'] == "Castanheiras"]
result = filtered_data['IDADEMAE'].agg(['min', 'max', 'mean'])
result
```

```python
# qual a idade media, maxima, minima dos pais nesse municipio?
result = filtered_data['IDADEPAI'].agg(['min', 'max', 'mean'])
result
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
data['DTNASC'] = pd.to_datetime(data['DTNASC'], errors='coerce')
marco_data = data[data['DTNASC'].dt.month == 3]
nascimentos_marco = marco_data['munResNome'].value_counts()
nascimentos_marco
```

```python
# qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
filtered_data = data[data['munResNome'] == "Porto Velho"]
result = filtered_data['QTDFILVIVO'].agg(['min', 'max', 'mean'])
result
```

```python
# qual a idade media, maxima, minima dos pais nesse municipio?
result = filtered_data['IDADEPAI'].agg(['min', 'max', 'mean'])
result
```

Vamos investigar como a quantidade de consultas pré-natais (CONSPRENAT) influencia os resultados do nascimento

```python
consultas_peso = data[['CONSPRENAT', 'PESO', 'SEMAGESTAC', 'APGAR1', 'APGAR5']].dropna()
stats_summary = consultas_peso.describe()
correlations = consultas_peso.corr()
stats_summary, correlations
```

Há uma correlação fraca positiva entre o número de consultas pré-natais e
Peso ao nascer (correlação: 0,067) e Idade gestacional (correlação: 0,0900.
Não há correlação significativa com os valores de Apgar no 1º e 5º minuto.
Embora as correlações sejam fracas, os dados indicam que o acompanhamento pré-natal pode ter impacto na saúde do bebê, especialmente em peso e idade gestacional.
