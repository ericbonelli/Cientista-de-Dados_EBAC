#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[(sinasc['APGAR5'] > 0) & (sinasc['APGAR5'] <= 3),
           'gravidade_afix'] = 'asfixia grave'
sinasc.loc[(sinasc['APGAR5'] > 4) & (sinasc['APGAR5'] <= 5),
           'gravidade_afix'] = 'asfixia moderada'
sinasc.loc[(sinasc['APGAR5'] > 6) & (sinasc['APGAR5'] <= 7),
           'gravidade_afix'] = 'asfixia leve'
sinasc.loc[(sinasc['APGAR5'] > 8) & (sinasc['APGAR5'] <= 10),
           'gravidade_afix'] = 'normal'
sinasc.head(5)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.countplot(x='GESTACAO', hue='gravidade_afix', data=sinasc)
plt.xticks(rotation=45)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
freq = pd.crosstab(sinasc['GESTACAO'], sinasc['gravidade_afix'])
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pctlin = freq.div(freq.sum(axis = 1), axis = 0)
freq_pctlin
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
freq_pctlin.plot.bar()
plt.xticks(rotation=70)
```

**Percebemos que acima de 37 semanas temos um percentual acima de 99% de nascimentos normais, na qual essa taxa vai diminuindo com o nascimento em semanas menores que 36 semanas, se agrvando em menos de 27 semanas, que atimgem o maior percentual de afixia.**
