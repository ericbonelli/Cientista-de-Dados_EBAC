```python
import pandas as pd
import numpy as np
import requests

# 1) seu código aqui
from google.colab import drive
drive.mount('/content/drive')
sinasc = pd.read_csv('/content/drive/MyDrive/Especialização em IA/Cientista de Dados - Ebac/Python em Ciência de Dados/Limpeza e preparação de dados/SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
```

```python
# 2) Conte o número de valores missing por variável.
sinasc.isna().sum()
```

```python
# 3) Ok, no item anterior você deve ter achado pouco prático ler a informação de tantas variáveis, muitas delas nem devem ser interesantes. Então crie uma seleção dessa base somente com as colunas que interessam. São elas: ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5'] Refaça a contagem de valores missings.
sinasc1 = sinasc[['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO',
                 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']]
sinasc1.isna().sum()
```

```python
sinasc1.shape
```

```python
# Apgar é uma nota que o pediatra dá ao bebê quando nasce de acordo com algumas características associadas principalmente à respiração. Apgar 1 e Apgar 5 são as notas 1 e 5 minutos do nascimento. Apgar5 será a nossa variável de interesse principal. Então remova todos os registros com Apgar5 não preenchido. Para esta seleção, conte novamente o número de linhas e o número de missings
sinasc1 = sinasc1.dropna(subset=['APGAR5'])
sinasc1.isna().sum()
```

```python
# 5) observe que as variáveis ['ESTCIVMAE', 'CONSULTAS'] possuem o código 9, que significa ignorado. Vamos assumir que o não preenchido é o mesmo que o código 9
sinasc1.fillna({'ESTCIVMAE': 9, 'CONSULTAS': 9}, inplace=True)
sinasc1.isna().sum()
```

```python
# 6)Substitua os valores faltantes da quantitativa (QTDFILVIVO) por zero
sinasc1.fillna({'QTDFILVIVO': 0}, inplace=True)
sinasc1.isna().sum()
```

```python
# 7) Das restantes, decida que valore te parece mais adequado (um 'não preenchido' ou um valor 'mais provável' como no item anterior) e preencha. Justifique. Lembre-se de que tratamento de dados é trabalho do cientista, e que estamos tomando decisões a todo o momento - não há necessariamente certo e errado aqui
# mostrar as categorias de gestacao
sinasc1['GESTACAO'].value_counts()
```

```python
sinasc1.fillna({'ESCMAE': '8 a 11 anos'}, inplace=True)
sinasc1.isna().sum()
```

**Resolveu-se preencher com 'ESCMAE': '8 a 11 anos' por ser a categoria com mais dados, trazendo um menor imapcto na proporcionalidade da amostra**

```python
# 8) O Apgar possui uma classificação indicando se o bebê passou por asfixia:
    #Entre 8 e 10 está em uma faixa 'normal'.
    #Entre 6 e 7, significa que o recém-nascido passou por 'asfixia leve'.
    #Entre 4 e 5 significa 'asfixia moderada'.
    #Entre 0 e 3 significa 'asfixia severa'.
    #Crie uma categorização dessa variável com essa codificação e calcule as frequências dessa categorização
sinasc1.loc[(sinasc1['APGAR5'] >= 8) & (sinasc1['APGAR5'] <= 10), 'APGAR5_CAT'] = 'normal'
sinasc1.loc[(sinasc1['APGAR5'] >= 6) & (sinasc1['APGAR5'] <=  7), 'APGAR5_CAT'] = 'asfixia leve'
sinasc1.loc[(sinasc1['APGAR5'] >= 4) & (sinasc1['APGAR5'] <=  5), 'APGAR5_CAT'] = 'asfixia moderada'
sinasc1.loc[(sinasc1['APGAR5'] >= 0) & (sinasc1['APGAR5'] <=  3), 'APGAR5_CAT'] = 'asfixia severa'
sinasc1['APGAR5_CAT'].value_counts()
```

```python
# Renomeie as variáveis para que fiquem no snake case, ou seja, em letras minúsculas, com um underscore entre as palávras. Dica: repare que se você não quiser criar um dataframe novo, você vai precisar usar a opção inplace = True.
sinasc1.columns = sinasc1.columns.str.lower()
sinasc1.head(5)
```
