### 5. Utilize a tabela do link abaixo e crie faixas utilizando o mapping e gere agrupamentos utilizando essas faixas como chave


### 5.1 IDH
A - https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_IDH-M


**Não sabia como pegar o dados na URL, então tive que pegar um codigo pronto de uma IA**

```python
import requests
from bs4 import BeautifulSoup

# Definir faixas de classificação do IDH
def classify_idh(idh):
    if idh < 0.550:
        return "IDH-M baixo"
    elif 0.550 <= idh < 0.700:
        return "IDH-M médio"
    elif 0.700 <= idh < 0.800:
        return "IDH-M alto"
    else:
        return "IDH-M muito alto"

# URL da página da Wikipedia
url = "https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_IDH-M"

# Fazer a requisição HTTP para obter o conteúdo da página
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")

    # Encontrar a tabela de IDH-M na página
    table = soup.find("table", {"class": "wikitable"})

    if table:
        # Criar o dicionário para armazenar os resultados
        idh_dict = {}

        # Iterar pelas linhas da tabela
        rows = table.find_all("tr")
        for row in rows[1:]:  # Pular o cabeçalho
            cols = row.find_all("td")
            if len(cols) >= 3:
                municipio = cols[1].text.strip()
                try:
                    idh = float(cols[2].text.strip().replace(",", "."))
                    idh_classification = classify_idh(idh)
                    idh_dict[municipio] = idh_classification
                except ValueError:
                    pass

        # Exibir o dicionário final
        print(idh_dict)
    else:
        print("Tabela de IDH-M não encontrada na página.")
else:
    print(f"Erro ao acessar a página: {response.status_code}")
```

```python
sinasc_raw = sinasc_raw.reset_index()
sinasc_raw = sinasc_raw.set_index('munResNome')
sinasc_raw.groupby(idh_dict).count()
```

### 5.3 PIB
C - https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_PIB


**Não sabia como pegar o dados na URL, então tive que pegar um codigo pronto de uma IA.**

```python
import requests
from bs4 import BeautifulSoup

# Definir faixas de classificação do PIB (valores exemplares)
def classify_pib(pib):
    if pib < 500_000:
        return "PIB muito baixo"
    elif 500_000 <= pib < 1_000_000:
        return "PIB baixo"
    elif 1_000_000 <= pib < 5_000_000:
        return "PIB moderado"
    else:
        return "PIB alto"

# URL da página da Wikipedia
url = "https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_PIB"

# Fazer a requisição HTTP para obter o conteúdo da página
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")

    # Encontrar a tabela de PIB na página
    tables = soup.find_all("table", {"class": "wikitable"}) # Corrected: find_all to get all tables

    if tables:
        # Criar o dicionário para armazenar os resultados
        pib_dict = {}

        # Iterate over each table
        for table in tables:
            rows = table.find_all("tr")
            # Iterate over rows of the current table
            for row in rows[1:]:  # Pular o cabeçalho
                cols = row.find_all("td")
                if len(cols) >= 4: # Corrected: ensure enough columns for municipio and pib
                    municipio = cols[2].text.strip() # Corrected: municipio is in the 3rd column (index 2)
                    pib_raw = cols[3].text.strip() # Corrected: pib_raw is in the 4th column (index 3)

                    # Depurar os valores lidos
                    #print(f"Processando: Município={municipio}, PIB={pib_raw}")

                    try:
                        pib = float(pib_raw.replace(' ', '').replace('.', ''))
                        pib_classification = classify_pib(pib)
                        pib_dict[municipio] = pib_classification
                    except ValueError:
                        print(f"Erro ao processar PIB para o município {municipio}")

        # Exibir o dicionário final
        print(pib_dict)
    else:
        print("Tabela de PIB não encontrada na página.")
else:
    print(f"Erro ao acessar a página: {response.status_code}")
```

```python
sinasc_raw = sinasc_raw.reset_index()
sinasc_raw = sinasc_raw.set_index('munResNome')
sinasc_raw.groupby(pib_dict).count()
```
