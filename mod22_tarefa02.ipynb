#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# ----- Monte um gráfico de linhas para as médias móveis dos ativos da sua base----

# 1. Retornos percentuais diários
yf_retornos = df_close.pct_change()

# 2. Média móvel de 90 dias dos retornos
yf_retornos_ma90 = yf_retornos.rolling(90).mean()

# 3. Transforma para formato "longo" (long format) para plotly express
yf_retornos_ma90_long = yf_retornos_ma90.stack().reset_index()
yf_retornos_ma90_long.columns = ['data', 'ativo', 'valor']

# 4. Gráfico interativo com Plotly Express
fig = px.line(yf_retornos_ma90_long, x="data", y="valor", color='ativo',
              title="Retorno Médio (90 dias) - Percentual Diário")
fig.show()
````

````python
# ----- Monte um gráfico de linhas para os desvios padrão em janela móvel-----

# 1. Retornos percentuais diários
yf_retornos = df_close.pct_change()

# 2. Desvio padrão (volatilidade) com janela de 90 dias
yf_volatilidade_ma90 = yf_retornos.rolling(90).std()

# 3. Transformar em formato longo para usar no plotly express
yf_volatilidade_ma90_long = yf_volatilidade_ma90.stack().reset_index()
yf_volatilidade_ma90_long.columns = ['data', 'ativo', 'valor']

# 4. Gráfico interativo
fig = px.line(yf_volatilidade_ma90_long, x="data", y="valor", color='ativo',
              title="Volatilidade (Desvio Padrão) - Janela de 90 dias")
fig.update_layout(yaxis_title="Volatilidade")
fig.show()
````

Períodos com picos no desvio padrão (volatilidade) devem coincidir com oscilações bruscas no gráfico de retorno.

Se a volatilidade sobe e o retorno médio cai, é sinal de crise, choque de mercado ou incerteza.

Se a volatilidade sobe mas o retorno médio sobe também, pode ser uma alta acelerada (euforia ou bolha).

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
from datetime import timedelta

lista_indicadores = ['Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['^GSPC']*len(lista_indicadores)
colunas = list(zip(lista_indicadores, ativo))

# Definir data final (hoje) e data inicial (60 dias corridos antes)
data_fim = pd.Timestamp.today()
data_ini = data_fim - timedelta(days=60)

# Recorte dos dados do yf_data
gspc = yf_data.loc[data_ini:data_fim, colunas].copy()
gspc.columns = lista_indicadores
gspc.head()
````

````python
graph = {
    'x': gspc.index,
    'open': gspc.Open,
    'close': gspc.Close,
    'high': gspc.High,
    'low': gspc.Low,
    'type': 'candlestick',
    'name': '^BVSP',
    'showlegend': True
}
````

````python
from plotly import graph_objects
from plotly.graph_objs import Layout

layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title="Índice S&P 500",
    layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
